```python
#dataset-2
dataset_2 = pd.read_excel(r"D:\NextHikes\project1\dataset_2.xlsx")
dataset_2.head()
```

```python
#dataset-3
dataset_3 = pd.read_excel(r"D:\NextHikes\project1\dataset_3.xlsx")
dataset_3.head()
```

```python
combine_data = pd.concat([dataset_1, dataset_2], ignore_index=True)
print(combine_data.head())
```

```python
concatenated_data = concatenated_data.fillna(0)
concatenated_data.info()

```

```python
from scipy.stats import zscore

# Assuming 'column_name' is the column where you want to detect outliers
z_scores = zscore(concatenated_data['weathersit'])
threshold = 3
print("z_scores ",z_scores )

# Indices of outliers
outlier_indices = (abs(z_scores) > threshold)
print("Indices of outliers",outlier_indices)

# Removing outliers
concatenated_data = concatenated_data[~outlier_indices]
print("Removing outliers",concatenated_data)
```
